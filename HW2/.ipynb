```python
import tensorflow as tf
import numpy as np
import pandas as pd
```

```python
ds = pd.read_csv("glass.csv",
                 names=["ID", "RI", "NA2O", "MGO", "AL2O3", "SIO2", "K2O", "CAO", "BAO", "FE2O3", "TYPE"])
```

```python
#ds
```

```python
a = ds.values
```

```python
features = a[:,1:10]
label = a[:,10]
type_glass = ds['TYPE']
b = pd.get_dummies(type_glass)
onehot = b.values
```

```python
def norm_data(input_data):
    input_data = np.float128(input_data)
    max_data = np.max(input_data)
    #max_data = float128(max_data)
    #print(max_data)
    min_data = np.min(input_data)
    #min_data = float128(min_data)
    #norm_data = (input_data - np.min(input_data)) / (np.max(input_data) - np.min(input_data))
    norm_data = np.true_divide(input_data - min_data,max_data - min_data)
    return norm_data
```

```python
features = norm_data(features)
```

```python
features
```

```python
model = tf.keras.models.Sequential([
#    tf.keras.layers.Dense(12, activation="relu",kernel_regularizer=tf.keras.regularizers.l2(0.001)),
#    tf.keras.layers.Dense(11, activation="relu",kernel_regularizer=tf.keras.regularizers.l2(0.001)),
#    tf.keras.layers.Dense(10,activation="relu"),
#    tf.keras.layers.Dense(9,activation="relu"),
#    tf.keras.layers.Dense(8,activation="relu"),
#    tf.keras.layers.Dense(7,activation="relu"),
#     tf.keras.layers.Dense(20,kernel_regularizer=tf.keras.regularizers.l2(0.0001),activation="relu"),
#     tf.keras.layers.Dropout(0.4),
#     tf.keras.layers.Dense(16,kernel_regularizer=tf.keras.regularizers.l2(0.0001),activation="relu"),
#     tf.keras.layers.Dropout(0.4),
    tf.keras.layers.Dense(70,kernel_regularizer=tf.keras.regularizers.l2(0.0001),activation="relu"),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.Dense(50,kernel_regularizer=tf.keras.regularizers.l2(0.0001),activation="relu"),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.Dense(40,kernel_regularizer=tf.keras.regularizers.l2(0.0001),activation="relu"),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.Dense(35,kernel_regularizer=tf.keras.regularizers.l2(0.0001),activation="relu"),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.Dense(30,kernel_regularizer=tf.keras.regularizers.l2(0.0001),activation="relu"),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.Dense(20,kernel_regularizer=tf.keras.regularizers.l2(0.0001),activation="relu"),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.Dense(15,kernel_regularizer=tf.keras.regularizers.l2(0.0001),activation="relu"),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.Dense(6),
    tf.keras.layers.Softmax()

])

loss = tf.keras.losses.CategoricalCrossentropy()

model.compile(loss=loss)

tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir='log')
```

```python
model.fit(features, onehot, epochs=3000, callbacks=[tensorboard_callback])
```

```python
model.evaluate(features, onehot)
```

```python
predict = model.predict(features)
# Calculate accuracy
out = np.argmax(predict,axis=1)
compare = out == label
accuracy = np.sum(compare) / 214 * 100
print(accuracy)
```
